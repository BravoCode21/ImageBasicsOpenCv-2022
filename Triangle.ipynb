```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\HP\Desktop\MVM\dog_backpack.JPG")
```

```python
pts = np.array([[150,200], [250,200], [200, 120]], np.int32)
```

```python
new_pts = pts.reshape((-1,1,2))
```

```python
new_pts
```

```python
isClosed = True
color = (0,255,0)
thickness = 5

#drawing polygon
new_im = cv2.polylines(img, [new_pts], isClosed, color, thickness)
```

```python
pts.shape
```

```python
plt.imshow(new_im)
```

```python
#fill in the triangle
```

```python
new_im = cv2.fillPoly(img, [new_pts], color)
plt.imshow(new_im)
```
