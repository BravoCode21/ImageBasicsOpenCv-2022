```python
#create a script that opens the image and its possible to click and create circles
```

```python
def draw_circle(event,x,y,flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x,y), 20, (0,255,0), 10)
        
cv2.namedWindow(winname = "drw_img")
cv2.setMouseCallback("drw_img", draw_circle)

while True:
    cv2.imshow("drw_img", img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
